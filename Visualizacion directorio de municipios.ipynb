# Librerias y dependencias

```python
#Manejo de datos
import pandas as pd

# Visualizacion de mapa
import folium
from folium.plugins import MarkerCluster
```

# Visualizacion de datos

```python
df = pd.read_csv('Datasets/Directorio_municipios_procesado.csv')

# Calcular el centro del mapa (promedio de lat y lon)
centro_lat = df['LATITUD'].mean()
centro_lon = df['LONGITUD'].mean()

mapa = folium.Map(location=[centro_lat, centro_lon], zoom_start=9)
cluster = MarkerCluster().add_to(mapa)


# Agregar marcadores para cada municipio
for _, row in df.iterrows():
    folium.Marker(
        location=[row['LATITUD'], row['LONGITUD']],
        popup=f"<b>Municipio:</b> {row['MUNICIPIO']}<br><b>Provincia:</b> {row['PROVINCIA']}",
        tooltip=row['MUNICIPIO']
    ).add_to(cluster)

# Mostrar mapa en Jupyter
mapa
```

## Descarga de mapa

```python
mapa.save("./Mapas/Mapa_directorio_de_municipios.html")
```
